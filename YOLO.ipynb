```python
models_to_train = [
    "yolov8n.pt",
    "yolov10n.pt",
    "yolo11n.pt"
]

results_summary = []

for model_name in models_to_train:

    print(f"\n===== Training {model_name} =====\n")

    model = YOLO(model_name)

    result = model.train(
        data="/content/dataset/data.yaml",
        epochs=50,
        imgsz=640,
        batch=16,
        project="yolo_research",
        name=model_name.replace(".pt", ""),
        plots=True,
        verbose=False
    )

    metrics = model.val()

    results_summary.append({
        "Model": model_name,
        "mAP50": metrics.box.map50,
        "mAP50-95": metrics.box.map,
        "Precision": metrics.box.mp,
        "Recall": metrics.box.mr
    })

    df = pd.DataFrame(results_summary)
    df.to_csv("yolo_model_comparison.csv", index=False)

```

```python
import pandas as pd

files = {
    "YOLOv11n": "/content/v11n_Results.csv",
    "YOLOv10n": "/content/v10n_results.csv",
    "YOLOv8n": "/content/v8n_Results.csv"
}

summary_list = []

for model_name, path in files.items():

    df = pd.read_csv(path)

    numeric_df = df.select_dtypes(include=['number'])

    averages = numeric_df.mean()

    averages["Model"] = model_name

    summary_list.append(averages)

comparison_df = pd.DataFrame(summary_list)

cols = ["Model"] + [col for col in comparison_df.columns if col != "Model"]
comparison_df = comparison_df[cols]

comparison_df.to_excel("YOLO_Model_Comparison.xlsx", index=False)

comparison_df
```
